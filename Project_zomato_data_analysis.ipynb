**Merging 2 files / data frames zomato dataset.csv and country-code.xlsx .**

```python
final_df = pd.merge(df,df2,on='Country Code',how='left')
```

```python
final_df.head(2)
```

**Datatypes of datasets .**

```python
final_df.dtypes
```

**how many records are present with respect to every country .**

```python
country_names = final_df.Country.value_counts().index
```

```python
country_vals = final_df.Country.value_counts().values
```

**Pie charts of top 3 countries**

```python
plt.title(" Top 3 countries in terms of orders ")

plt.pie(country_vals[:3],labels = country_names[:3],
        autopct= '%1.2f%%',explode=[0,0.1,0.2])
plt.show()
```

**Observations :**

* Zomato operation mostly focused in india  and not in other countries .

**checking Rating .**

```python
ratings = final_df.groupby(
    ['Aggregate rating', 'Rating color', 'Rating text']
                 ).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

**Rating_Observations**

* Rating between 4.5 to 5 --> 
Excellent 
* Rating between 3.4 to 4.5 --> 
very good
* Rating between 3.5 to 3.9 -->
good 
* Rating between 3.0 to 3.4 --> 
average
* Rating between 1.8 to 2.4 --> 
poor


```python
# increase size of graph
plt.title("Bar graph Aggregate-rating by Rating-count ")

plt .rcParams['figure.figsize']=(12,6)

sns.barplot(x='Aggregate rating', y='Rating count',
            data = ratings)
```

```python
plt.title("Bar graph Aggregate-rating by Rating-count ")

sns.barplot(x='Aggregate rating', y='Rating count',
            hue= 'Rating color',data = ratings,
            palette = ['blue','red','orange','yellow','green','green'])
```

**Onservation_3**

* Not Rated count is very high
* Maximum number of rating is between 2.7 to 3.9
